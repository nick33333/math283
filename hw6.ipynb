```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def func_q1c(mu, n, xbar1=31, xbar2=19):
    z1 = (xbar1-mu)/(12/np.sqrt(n))
    z2 = (xbar2-mu)/(12/np.sqrt(n))
    return norm.cdf(z1) - norm.cdf(z2) 
```

```python
mus = np.arange(0, 51, 1)
p_vals = [func_q1c(mu=mu, n=9, xbar1=31.5794, xbar2=18.4206) for mu in mus]
plt.plot(mus, p_vals, label="n = 9")
plt.xlabel(r"$\mu_{1}$")
plt.ylabel(r"$\beta$")

p_vals = [func_q1c(mu=mu, n=16, xbar1=29.93456, xbar2=20.06544) for mu in mus]
plt.plot(mus, p_vals, label="n = 16")
plt.xlabel(r"$\mu_{1}$")
plt.ylabel(r"$\beta$")

plt.suptitle(r"H-601 (c) Type II Error VS $\mu_{1}$")
# plt.title(r"($\bar{x}$=31)")

plt.legend()
```

```python
mus[np.argmax(p_vals)]
```

```python
mus = np.arange(0, 51, 1)
p_vals = [func_q1c(mu=mu, n=9, xbar1=31.5794, xbar2=18.4206)  for mu in mus]
plt.plot(mus, 1-np.array(p_vals), label="n = 9")
plt.xlabel(r"$\mu_{1}$")
plt.ylabel(r"$\beta$")

p_vals = [func_q1c(mu=mu, n=16, xbar1=29.93456, xbar2=20.06544) for mu in mus]
plt.plot(mus, 1-np.array(p_vals), label="n = 16")
plt.xlabel(r"$\mu_{1}$")
plt.ylabel(r"1 - $\beta$")

plt.suptitle(r"H-601 (c) Power Curve $\mu_{1}$")

plt.legend()
```

```python
mus[np.argmin(1-np.array(p_vals))]
```
